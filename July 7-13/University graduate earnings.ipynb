# Median annualised earnings by subject. Is there a difference between famale and male?

```python
df.groupby(by = 'subject').medianAnnEarn.median().sort_values(ascending=False)
```

```python
df.groupby(by = 'subject').medianAnnEarn.median().sort_values().\
plot(x='medianAnnEarn', y='subject', kind='barh', figsize=(8,8))
```

```python
df[(df.sex == 'Male') & (df.yearsAfterGraduation == 5)].groupby(by = 'subject').medianAnnEarn.median().sort_values(ascending=False)
```

```python
df[(df.sex == 'Female') & (df.yearsAfterGraduation == 5)].groupby(by = 'subject').medianAnnEarn.median().sort_values(ascending=False)
```

```python
male_median = df[(df.sex == 'Male') & (df.yearsAfterGraduation == 5)].groupby(by = 'subject').\
medianAnnEarn.median().sort_values(ascending=False)
female_median = df[(df.sex == 'Female') & (df.yearsAfterGraduation == 5)].groupby(by = 'subject').\
medianAnnEarn.median().sort_values(ascending=False)
subjects = []
for subject in df.subject:
    if subject not in subjects:
        subjects.append(subject)
subjects

data = pd.DataFrame({'male': male_median,'female':female_median, 'subject': subjects})
data
```

```python
ax = data.sort_values(by='subject', ascending=False).plot(x='subject', y=['male', 'female'], kind='barh', 
                                                          width=0.5, figsize=(10, 10), color=['steelblue', 'orange'])
ax.xaxis.tick_top()
```

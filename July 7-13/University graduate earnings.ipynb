# Groups with the highest / lowest median annualised earnings

```python
df.sort_values(by='medianAnnEarn', ascending=False).head(10)
```

```python
df[df.medianAnnEarn > 0].sort_values(by='medianAnnEarn', ascending=False).tail(15)
```

# Comparison of median annualised earnings: males vs. female one / three / five years after graduation

```python
df[df.yearsAfterGraduation == 1].groupby(by = 'sex').medianAnnEarn.median()
```

```python
ax = df[(df.yearsAfterGraduation == 1) & (df.sex == 'Male')].medianAnnEarn.hist(bins=30)
df[(df.yearsAfterGraduation == 1) & (df.sex == 'Female')].medianAnnEarn.hist(ax=ax, alpha=0.5, bins=30)
```

```python
df[df.yearsAfterGraduation == 3].groupby(by = 'sex').medianAnnEarn.median()
```

```python
ax = df[(df.yearsAfterGraduation == 3) & (df.sex == 'Male')].medianAnnEarn.hist(bins=30)
df[(df.yearsAfterGraduation == 3) & (df.sex == 'Female')].medianAnnEarn.hist(ax=ax, alpha=0.5, bins=30)
```

```python
df[df.yearsAfterGraduation == 5].groupby(by = 'sex').medianAnnEarn.median()
```

```python
ax = df[(df.yearsAfterGraduation == 5) & (df.sex == 'Male')].medianAnnEarn.hist(bins=30)
df[(df.yearsAfterGraduation == 5) & (df.sex == 'Female')].medianAnnEarn.hist(ax=ax, alpha=0.5, bins=30)
```

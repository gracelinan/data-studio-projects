# Median annualised earnings comparison one, three and five years after graduation

```python
df.groupby(by = 'yearsAfterGraduation').medianAnnEarn.median()
```

```python
df[df.yearsAfterGraduation == 1].sort_values(by='medianAnnEarn', ascending=False).head()
```

```python
one = df[df.yearsAfterGraduation == 1][['UKPRN', 'providerName', 'Region', 'subject', 'sex', 'medianAnnEarn']]
one['oneMedian'] = one['medianAnnEarn'] 
three = df[df.yearsAfterGraduation == 3][['UKPRN','subject', 'sex', 'medianAnnEarn']]
three['threeMedian'] = three['medianAnnEarn'] 
merged = pd.merge(one, three, on=['UKPRN', 'subject', 'sex'])
merged.head()
```

```python
five = df[df.yearsAfterGraduation == 5][['UKPRN','subject', 'sex', 'medianAnnEarn']]
five['fiveMedian'] = five['medianAnnEarn'] 
year = pd.merge(merged, five, on=['UKPRN', 'subject', 'sex']).\
drop(columns=['medianAnnEarn_x', 'medianAnnEarn_y', 'medianAnnEarn'])
```

```python
year['growth'] = (year.fiveMedian - year.oneMedian) / year.oneMedian * 100
year.sort_values(by='growth', ascending=False)
```

```python
year.groupby(by = 'sex').growth.median()
```
